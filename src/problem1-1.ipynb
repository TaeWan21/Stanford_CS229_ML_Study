```python
from p01b_logreg import LogisticRegression
from p01e_gda import GDA
import util
import numpy as np

train_path = '/Users/kimtaewan/Desktop/개인공부/ML_440/code/data/ds2_train.csv'
eval_path = '/Users/kimtaewan/Desktop/개인공부/ML_440/code/data/ds2_valid.csv'

x_train, y_train = util.load_dataset(train_path, add_intercept=True)
x_valid, y_valid = util.load_dataset(eval_path, add_intercept=True)
# Logistic Regression
logistic_reg = LogisticRegression()
logistic_reg.fit(x_train, y_train)

# GDA
gda = GDA()
gda.fit(x_train, y_train)


# training set 의 결정경계 Plot
util.plot(x_train, y_train, logistic_reg.theta, gda.theta, save_path='res/LR_GDA_train_DS2_plot.png')
util.plot(x_valid, y_valid, logistic_reg.theta, gda.theta, save_path='res/LR_GDA_valid_DS2_plot.png')
```
